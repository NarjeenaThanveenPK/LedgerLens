# Financial Data Analysis

This notebook analyzes financial data extracted from the 2023–2025 10K reports of Microsoft, Apple, and Tesla using Python and pandas.

## Methodology

- Extracted Total Revenue, Net Income, Total Assets, Total Liabilities, and Operating Cash Flow from SEC 10K reports.
- Stored the data in CSV format.
- Loaded the dataset using pandas.
- Calculated Revenue Growth (%) and Net Income Growth (%) using year-over-year percentage change.
- Used descriptive statistics and group-by operations to compare company performance.

```python
import pandas as pd
```

```python
df = pd.read_csv('D:/Projects/GenAI/financial_data.csv')
df.head(10)
```

```python
df = df.sort_values(["Company","Year"])
df.info()
```

```python
col = ["Total Revenue","Net Income","Total Assets","Total Liabilities","Operating Cash Flow"]
for i in col:
    df[i]=(
        df[i].astype(str)
        .str.replace(",","")
        .astype(float)
    )

df.info()
```

```python
df.describe()
```

```python
df["Revenue Growth (%)"]=df.groupby(["Company"])["Total Revenue"].pct_change()*100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100

df
```

```python
df.groupby("Company")["Total Revenue"].mean()
```

```python
df.groupby("Company")["Net Income"].mean()
```

```python
df.groupby("Year")[["Total Revenue", "Net Income"]].sum()
```

## Findings

- Apple recorded the highest average revenue and net income during the analysis period.
- Microsoft demonstrated consistent revenue and profit growth across all three years.
- Tesla generated positive revenue but experienced a decline in both revenue growth and net income during 2025.
- The combined revenue of Microsoft, Apple, and Tesla increased from 2023 to 2025, indicating overall business growth.
- Revenue Growth (%) and Net Income Growth (%) provide useful indicators for evaluating company performance over time.

## Conclusion

The extracted financial metrics and calculated growth rates provide useful insights that could support the development of an AI-powered financial chatbot capable of answering questions about company performance.
